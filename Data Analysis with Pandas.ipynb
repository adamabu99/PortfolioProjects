PERFORM TEXT DATA CLEANING BY REMOVING STOPWORDS

```python
!pip install nltk
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stopwords.words('english')
```

```python
stop_words = stopwords.words('english')
stop_words.extend(['Amazon', 'amazon', 'alexa', 'echo', 'Alexa', 'device','dot','Dot'])
```

```python
!pip install gensim
```

```python
import gensim
from gensim.utils import simple_preprocess
```

```python
gensim.utils.simple_preprocess(df['verified_reviews'][0])
```

```python
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if len(token) >= 1 and token not in stop_words:
            result.append(token)
            
    return result
```

```python
df['no_stopwords'] =df['verifeid_reviews_nonpunc'].apply(preprocess)
```

```python
df
```

```python
df['joined'] = df['no_stopwords'].apply(lambda x: " ".join(x) )
```
